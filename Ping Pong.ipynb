```python
import pygame , sys
import random
from pygame import display, event
pygame.init()

#creating screen
screen = display.set_mode((600,400))
display.set_caption("Ping Pong")

#variables
screen_height = 400
screen_width = 600

player_x = 0
player_y = 150
player_width = 8
player_height = 70
p_vel = 8

enemy_x = 592
enemy_y = 150
enemy_width = 8 
enemy_height = 70
e_vel = 7

ball_x = 286
ball_y = 190
b_velx = 9
b_vely = 9
ball_width = 30
ball_height = 20

myfont = pygame.font.SysFont("monospace", 16)
score = 0

def creating_objects():
    pygame.draw.aalines(screen, (255,245,230), True, [(300,0),(300,400)])
    pygame.draw.ellipse(screen, (255,245,230), (ball_x,ball_y,30,20))
    pygame.draw.rect(screen, (255,245,230), (player_x,player_y,player_width,player_height))
    pygame.draw.rect(screen, (255,245,230), (enemy_x,enemy_y,enemy_width,enemy_height))
    pygame.draw.aalines(screen, (178,34,34), True, [(0,0),(600,0),(600,400),(0,400)])
    pygame.draw.aalines(screen, (178,34,34), True, [(599,0),(599,399),(0,399),(0,0)])

def restart():
    global ball_x,ball_y,score,b_velx,b_vely
    ball_x = 290
    ball_y = 190
    b_velx *= random.choice((1,-1))
    b_vely *= random.choice((1,-1))
    score = 0
    
def movement_of_ball():
    global b_velx,score,b_vely,ball_x,ball_y
    ball_x += b_velx
    ball_y += b_vely
    if ball_x <= 0 or ball_x >= screen_width-24:
        restart()
    if ball_y <= 1 or ball_y > screen_height-24:
        b_vely *= -1
    if player_x + player_width >= ball_x and player_x <= ball_x + ball_width and player_y + player_height >= ball_y and player_y <= ball_y + ball_height:
        b_velx *= -1
        score = score + 1
    if ball_x + ball_width >= enemy_x and ball_x <= enemy_x + enemy_width and ball_y + ball_height > enemy_y and ball_y <= enemy_y + enemy_height:
        b_velx *= -1

def movement_of_enemy():
    global enemy_y
    if enemy_y < ball_y: 
        enemy_y += e_vel
    if enemy_y > ball_y:
        enemy_y -= e_vel
    if enemy_y < 0:
        enemy_y = 0
    if enemy_y > screen_height-enemy_height:
        enemy_y = screen_height-enemy_height
    
#main loop
run = True
while run:
    pygame.time.delay(60)
    screen.fill((61,89,171))
    
    #till when the game will run
    for events in event.get():
        if events.type == pygame.QUIT:
            run = False
            
    keys = pygame.key.get_pressed()
    
    #movement of player
    if keys[pygame.K_UP] and player_y >= p_vel:
        player_y -= p_vel
    if keys[pygame.K_DOWN] and player_y <= screen_height-p_vel-player_height:
        player_y += p_vel    
    
    movement_of_ball()
    movement_of_enemy()
    
    scoretext = myfont.render("Score = "+str(score), 1, (240,128,128))
    screen.blit(scoretext, (5, 10))
    creating_objects()
    display.update()  
    
pygame.quit()    
```
